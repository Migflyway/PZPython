### 开始
要开始处理客户数据，我们首先需要  `import` 所需的功能，并将数据导入 pandas。

利用DataFrame中的函数运行以下代码单元格，加载我们的数据并使用 `.head(15) ` 函数显示前几个客户以进行检查。
> **提示：**你可以通过点击单元格和使用键盘快捷键 **Shift + Enter** 或 **Shift + Return** 运行代码单元格。此外，在选中单元格后还可以使用工具栏中的**播放**按钮执行代码单元格。双击 Markdown 单元格（如下所示的文本单元格）即可进行编辑，并使用相同的快捷键保存。

````python
#Import Numpy and Pandas
import numpy as np
import pandas as pd
from IPython.display import display   # Allows the use of display() for DataFrames

# Pretty display for notebooks
%matplotlib inline

# TO DO: Load the data to 'full_data'
full_data = pd.read_excel('Investors.xlsx')

# Print the first few rows of the data
display(full_data.head(15))
````

让我们来仔细看一下，这个客户数据表格的内容：
- **姓名：** 客户的中文姓名。有重名，例如编号4和编号编号11，姓名相同，但是年龄不同，账号不同，是两个不同的人
- **性别：** 客户的性别（男/女）
- **年龄：** 客户的年龄
- **投资市场：** 客户投资的市场（A股/港股/美股）
- **投资账户号码：** 客户的账户号码，其中A-A股，H-港股，U-美股
- **开户时间：** 客户开户的具体日期
- **投资收益：** 表示开户至今（2018年9月1日），客户的累计收益/损失
- **资产余额：** 表示客户在本时点（2018年9月1日），账户中所有资产的余额

## 一些描述性的统计

### 问题1（4分）：
我们先来尝试回答这几个简单的问题：
1. 投资者中，年龄最大的是几岁？
2. 开户时间最早的是哪一天开户的？
3. 所有投资者的资产余额，平均数是多少？中位数又是多少？（保留2位小数）

>**提示：** numpy中有min, max等常用的函数可以使用。 print 资产余额时注意调整打印格式(可以参考Week2挑战题第一题所提到的方法)

````python
# 将下列''替换为你的代码
# TO DO: Print Largest Age among all investors
largest_age = np.max(full_data['年龄'])

# TO DO: Print Earliest Account Open Date
earliest_open_date = np.min(full_data['开户时间'])

# TO DO: Print Average & Median of Account Balance
average_account_balance = np.mean(full_data['资产余额'])
median_account_balance = np.median(full_data['资产余额'])

# Print result
print('Largest Age among all investors is', largest_age)
print('Earliest Account Open Date is', earliest_open_date)
print('Average Account Balance is', "%.2f" % average_account_balance)
print('Median Account Balance is', "%.2f" % median_account_balance)
````

**最后打印结果应该为**
- Largest Age among all investors is 58
- Earliest Account Open Date is 2016-01-17 00:00:00
- Average Account Balance is 146886.05
- Median Account Balance is 140649.78

除了以上的方法，我们还可以直接用describe来查看数据的统计性质。
请使用describe来查看，并自行检验上一步骤的运算结果是否正确。

````python
# TO DO: Print Description of full_data’
print(full_data['年龄'].describe())
print(full_data['开户时间'].describe())
print(full_data['资产余额'].describe())
````

## 进一步的处理
我们已经知道，重名的投资者很多。我们现在来尝试创建一个新的列作为唯一的key

### 问题2（1分）：
创建新的一列'new'，这一列的值为 姓名 + 投资账户，要求更新原full_data，然后用display和`head()`命令，显示数据的前几行

````python
# TO DO: Create 'new' column, Value = Name + A/C No.
'''请输入你的代码'''
full_data['new'] = full_data['姓名'] + full_data['投资账户']
display(full_data.head(15))
````

### 问题3（2分）：
- 创建新的一列'Aging'，这一列的值为：报告日期(report_date: 2018-09-01)减去开户时间，表示该客户开户至今已有多长时间。
- 要求更新原full_data，然后用display和`head()`命令，显示数据的前几行

````python
import datetime
report_date = datetime.datetime.strptime('2018-09-01', '%Y-%m-%d') #对report_date进行数据格式转换

# TO DO: Create 'Aging' column, Value = report_date - A/C Open Date
full_data['Aging'] = report_date - full_data['开户时间']
display(full_data.head(15))
````

### 问题4（2分）：
我们原来的列名都是中文，现在我们想要把所有中文列名都改成英文。

1. 请根据给出的名称更改规则 rename_dic，对full_data的列名进行重命名，要求直接更新 full_data
2. 请列出重命名的这些列的数据，并用display和`head()`显示前几行

>**提示：** `df.rename()`更改列名，`dictionary.values()`可获得所有字典所有key对应的值, `list()`可以把其他对象转换成一个列表

````python
# Rename rules as dictionary
rename_dic = {'姓名': 'Name',
             '性别': 'Gender',
             '年龄': 'Age',
             '投资账户': 'AC_No',
             '开户时间': 'AC_Open',
             '投资收益': 'Gain_Loss',
             '资产余额': 'AC_Balance'}
list(rename_dic.values())
````

````python
# TO DO: Rename columns according to rename fules
# Then display renamed columns with head()
'''请输入你的代码'''
for key in rename_dic:
    full_data.rename(columns={key:rename_dic[key]}, inplace=True)
        
display(full_data.head(15))
````

### 问题5（2分）：
我们可以根据投资者过往的投资表现，把客户做一个简单的区分。对于那些投资收益好的客户，我们把他们认作“优秀投资者”

* 新增一列'type'，对于投资收益大于等于3万的客户，记录为'优秀投资者'，其余为'普通投资者'
* 并用display和`head()`显示 full_data 的前几行

>**要求：** 
1. 利用`df.loc[]`定位
2. 利用“`~`”符号对条件取非（条件：`full_data['Gain_Loss']>=30000`）

>**提示：**
- ~符号python常用的逻辑运算符之一，表示反运算。
- `df.loc['筛选条件', '新增列名'] = '' `<--通过这种方式可以赋值给新增列

````python
# TO DO: Mark '优秀投资者' given Gain_Loss >=30000, otherwise mark '普通投资者'
'''请输入你的代码'''
full_data.loc[full_data['Gain_Loss'] >= 30000, 'type'] = '优秀投资者'
full_data.loc[~(full_data['Gain_Loss'] >= 30000), 'type'] = '普通投资者'
display(full_data.head(15))
````

### 问题6（2分）：
对于上一步标注为'优秀投资者'的客户，我们来计算一下他们资产余额的平均值

* 计算'优秀投资者'的资产余额平均值，并且print出来。（仅一行代码）

>**提示：** `df.loc[]`定位

````python
# TO DO: Calculate average AC_Balance
'''请输入你的代码'''
np.mean(full_data.loc[full_data['type']=='优秀投资者','AC_Balance'])
````
